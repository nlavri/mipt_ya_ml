```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("coursera_sessions_train.txt", delimiter=';', header=None, dtype=str)
#s1 = str.join(',', ['59', '60', '61', '62', '60', '63', '64', '65', '66', '61', '67', '68', '67'])
#s2 = str.join(',',['67','60', '63'])
#train = pd.DataFrame([[s1, s2]])
train['view'] = train[0]
train['buy'] = train[1]
```

```python
train
```

```python

view_arr = []
for row in train.view:
    ids = row.split(',')
    for idvalstr in ids:
        view_arr.append(int(idvalstr))
        
buy_arr = []
for row in train.buy:
    if type(row) is str:
        ids = row.split(',')
        for idvalstr in ids:
            buy_arr.append(int(idvalstr))
       
```

```python
from collections import Counter
from collections import OrderedDict

class OrderedCounter(Counter, OrderedDict):
     'Counter that remembers the order elements are first encountered'

     def __repr__(self):
         return '%s(%r)' % (self.__class__.__name__, OrderedDict(self))

     def __reduce__(self):
         return self.__class__, (OrderedDict(self),)
```

```python
counter_view=Counter(view_arr)
counter_buy=Counter(buy_arr)
```

```python
#AverageRecall@1, AveragePrecision@1, AverageRecall@5, AveragePrecision@5
```

```python
def get_metrics(data, hist, k):
    kf = float(k)
    recall = []
    precision = []
    for index, row in data.dropna().iterrows():
        ids = row[0].split(',')
        idvals = map(int, ids)
        freqs = OrderedCounter()
        #print idvals
        for idval in idvals:
            #if idval in hist.index:
                #freqs[idval] = hist.loc[idval][1]
                #print idval, hist[idval]
                freqs[idval] = hist[idval]
            #else:
            #    freqs[idval] = 0
        #s_vals = OrderedC(sorted(freqs.items(), key=lambda t: -t[1]))
        #print freqs
        s_vals_pairs = freqs.most_common(k)# OrderedCounter(freqs.values)
        #print s_vals_pairs
        s_vals = map(lambda x: x[0],s_vals_pairs)
        #print s_vals
        #print ''
        #print 'raw ', freqs.items()
        #print  'sort ', s_vals
        b_ids = row[1].split(',')
        b_idvals = map(int, b_ids)
        #print 'rec ', s_vals.keys()[:k]
        #print 'buy ', b_idvals
        #print np.unique(s_vals)
        inter = np.intersect1d(s_vals, b_idvals)
        #print '-------'
        #print inter
        rec = float(inter.shape[0]) / float(len(b_ids))
        #print rec
        recall.append(rec)
        
        prec = float(inter.shape[0]) / kf #min(kf, len(idvals))
        #print prec
        precision.append(prec)
        
    return (round(np.mean(recall), 2), round(np.mean(precision), 2))
```

```python
vm1 = get_metrics(train, counter_view, 1)
print vm1
vm2 = get_metrics(train, counter_view, 5)
print vm2
bm1 = get_metrics(train, counter_buy, 1)
print bm1
bm2 = get_metrics(train, counter_buy, 5)
print bm2
```
