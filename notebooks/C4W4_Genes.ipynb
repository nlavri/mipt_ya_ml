```python
import pandas as pd
import numpy as np
import scipy.stats
import statsmodels
import statsmodels.sandbox.stats.multicomp
```

```python
def write_answer_to_file(answer, filename):
    with open(filename, 'w') as f_out:
        f_out.write(str(answer))
        
data = pd.read_csv("gene_high_throughput_sequencing.csv")
```

```python
normal = data[data['Diagnosis'] == 'normal']
early_neo = data[data['Diagnosis'] == 'early neoplasia']
cancer = data[data['Diagnosis'] == 'cancer']
```

```python
def F(c,t):
    return t / c if t > c else -c / t
```

```python
st1 = scipy.stats.ttest_ind(normal.drop(['Patient_id', 'Diagnosis'],axis=1), early_neo.drop(['Patient_id', 'Diagnosis'],axis=1), axis=0, equal_var = False)
v1 = sum(st1.pvalue < 0.05)
print v1 

write_answer_to_file(v1, "c4w4_1.txt")

st2 = scipy.stats.ttest_ind(early_neo.drop(['Patient_id', 'Diagnosis'],axis=1), cancer.drop(['Patient_id', 'Diagnosis'],axis=1), axis=0, equal_var = False)
v2 = sum(st2.pvalue < 0.05)
print v2 

write_answer_to_file(v2, "c4w4_2.txt")

normal.iloc[:,i]
```

```python
st3 = scipy.stats.ttest_ind(normal.drop(['Patient_id', 'Diagnosis'],axis=1), early_neo.drop(['Patient_id', 'Diagnosis'],axis=1), axis=0, equal_var = False)

reject, p_corrected1, a1, a2 = statsmodels.sandbox.stats.multicomp.multipletests(st3.pvalue, 
                                            alpha = 0.025, 
                                            method = 'holm') 
print sum(reject)
c1 = np.where(p_corrected1 < 0.025)
f1 = []
for i in c1[0]:
    f = F(normal.iloc[:,i+2].mean(), early_neo.iloc[:,i+2].mean())
    if(abs(f) > 1.5):
        f1.append(f1)
print len(f1)        
write_answer_to_file(len(f1), "c4w4_3.txt")

st4 = scipy.stats.ttest_ind(early_neo.drop(['Patient_id', 'Diagnosis'],axis=1), cancer.drop(['Patient_id', 'Diagnosis'],axis=1), axis=0, equal_var = False)
reject, p_corrected2, a1, a2 = statsmodels.sandbox.stats.multicomp.multipletests(st4.pvalue, 
                                            alpha = 0.025, 
                                            method = 'holm') 
print sum(reject)
c2 = np.where(p_corrected2 < 0.025)
f2 = []
for i in c2[0]:
    f = F(early_neo.iloc[:,i+2].mean(), cancer.iloc[:,i+2].mean())
    if(abs(f) > 1.5):
        f2.append(f2)
print len(f2)        
write_answer_to_file(len(f2), "c4w4_4.txt")
```
